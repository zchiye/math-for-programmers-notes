```python
# 2.43regular_polygon(n)[]
def regular_polygon(n):
    angle = 2 * pi / n
    return [to_cartesian((1, i * angle)) for i in range(n)]

draw(Polygon(*regular_polygon(5), color='blue'),
    # Polygon(*regular_polygon(6), color='red'), 
    Polygon(*regular_polygon(7), color='green'),
    # Polygon(*regular_polygon(8), color='purple')
)
```
