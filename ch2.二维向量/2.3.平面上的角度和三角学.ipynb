```python
# 练习2.41: 龙嘴巴的夹角是多少？脚趾的夹角是多少？尾巴的夹角是多少？
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]

mouth = [subtract((-5,2), (-2,2)), subtract((-2,2), (-5,1))]
toe = [subtract((0,-3), (-1,-4)), subtract((-1,-4), (1,-4))]
tail = [subtract((6,4), (3,1)), subtract((6,4), (5,1))]

def angle_between(v1, v2):
    polar1 = to_polar(v1)
    polar2 = to_polar(v2)
    angle = abs(polar1[1] - polar2[1]) * 180 / pi
    if angle > 90:
        angle = 180 - angle
    return angle

print("mouth angle =", angle_between(*mouth))
print("toe angle =", angle_between(*toe))
print("tail angle =", angle_between(*tail))
```
