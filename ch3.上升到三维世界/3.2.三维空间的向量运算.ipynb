```python
# n维向量加法
def add(*vectors):
    zipped = zip(*vectors)
    return tuple([sum(vector) for vector in zipped])

v1 = (1,1,3,1)
v2 = (2,4,-4,2)
v3 = (4,2,-2,3)
v4 = (1,1,1,4)
v5 = add(v1, v2, v3, v4)
print(v5)  # (8, 8, -2)
```

```python
# 向量减法
def subtract(v, w):
    return add(v, tuple([-x for x in w]))

v6 = subtract(v5, v1)
print(v6)
```

```python
# n维向量长度
from math import sqrt
def length(vector):
    return sqrt(sum([x**2 for x in vector]))

print(length(v5))
```

```python
# 向量距离
def distance(v, w):
    return length(subtract(v, w))

print(distance(v1, v2))
print(length((1,3,-7,1)))
```

```python
# 练习3.9: 找出另一组三个整数，以它们为坐标定义的向量也有整数长度。
def find_integer_vector(max):
    for x in range(1, max):
        for y in range(1, max):
            for z in range(1, max):
                if length((x,y,z)).is_integer():
                    yield (x,y,z)

for v in find_integer_vector(20):
    print(v, int(length(v)))
```

```python
# 练习3.10：找到一个与(-1, -1, 2)方向相同但长度为1的向量
def same_direction(v, target):
    return tuple([x/length(v)*target for x in v])
v = (-1, -1, 2)
print(same_direction(v, 1))  # (-0.4082482904638631, -0.4082482904638631, 0.8164965809277261)
```
